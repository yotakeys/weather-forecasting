```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.info()
```

```python
train = train.drop('winddirection_10m_dominant (°)', axis=1)
train.info()
```

```python
train.dropna(inplace=True)
train.isna().sum()
```

```python
train.drop('time', axis=1, inplace=True)
```

```python
train.head()
```

```python
#ENCODING STRING DATASET

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['sunrise (iso8601)'] = le.fit_transform(train['sunrise (iso8601)'])
train['sunset (iso8601)'] = le.fit_transform(train['sunset (iso8601)'])
train['city'] = le.fit_transform(train['city'])
```

```python
train.head()
```

```python
train['sunrise (iso8601)'].nunique()
```

```python
X = pd.DataFrame(train.iloc[:, :])
X = X.drop('rain_sum (mm)', axis=1)
y = pd.DataFrame(train.iloc[:,8])
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn import metrics
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)
print('nilai MSE : ', metrics.mean_squared_error(y_test, predictions))
```
